```python
%pip install -i https://pypi.gurobi.com gurobipy
import gurobipy as gp
from gurobipy import GRB
```

```python
m = gp.Model("Transbordo")

x_13 = m.addVar(vtype=GRB.CONTINUOUS, lb=0, name="Transporte_1_3")
x_14 = m.addVar(vtype=GRB.CONTINUOUS, lb=0, name="Transporte_1_4")
x_23 = m.addVar(vtype=GRB.CONTINUOUS, lb=0, name="Transporte_2_3")
x_24 = m.addVar(vtype=GRB.CONTINUOUS, lb=0, name="Transporte_2_4")
x_35 = m.addVar(vtype=GRB.CONTINUOUS, lb=0, name="Transporte_3_5")
x_36 = m.addVar(vtype=GRB.CONTINUOUS, lb=0, name="Transporte_3_6")
x_37 = m.addVar(vtype=GRB.CONTINUOUS, lb=0, name="Transporte_3_7")
x_45 = m.addVar(vtype=GRB.CONTINUOUS, lb=0, name="Transporte_4_5")
x_46 = m.addVar(vtype=GRB.CONTINUOUS, lb=0, name="Transporte_4_6")
x_47 = m.addVar(vtype=GRB.CONTINUOUS, lb=0, name="Transporte_4_7")

m.setObjective(1*x_13 + 3*x_14 + 1*x_23 +
               2*x_24 + 1*x_35 + 3*x_36 +
               4*x_37 + 3*x_45 + 3*x_46 +1*x_47,
               GRB.MINIMIZE)

m.addConstr(x_13 + x_14  <= 800,
            name="Capacidade Centro 1")
m.addConstr(x_23 + x_24  <= 1000,
            name="Capacidade Centro 2")
m.addConstr(x_35 + x_45 == 500, name="Demanda Mercado 1")
m.addConstr(x_36 + x_46 == 400, name="Demanda Mercado 2")
m.addConstr(x_37 + x_47 == 900, name="Demanda Mercado 3")

#Restricoes Transbordo
m.addConstr(x_13 + x_23 -x_35- x_36 - x_37 == 0, name="Tranbordo1")
m.addConstr(x_14 + x_24 -x_45- x_46 - x_47 == 0, name="Tranbordo1")

m.optimize()
```

```python
print("Custo mínimo: "+str(m.objVal))
print("Quantidade enviada:")
print("Origem\tDestino\tQuantidade")
print("1\t3\t"+str(x_13.X))
print("1\t4\t"+str(x_14.X))
print("2\t3\t"+str(x_23.X))
print("2\t4\t"+str(x_24.X))
print("3\t5\t"+str(x_35.X))
print("3\t6\t"+str(x_36.X))
print("3\t7\t"+str(x_37.X))
print("4\t5\t"+str(x_45.X))
print("4\t6\t"+str(x_46.X))
print("4\t7\t"+str(x_47.X))
```
